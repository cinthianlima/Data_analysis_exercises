### 5) Binomial and Poisson distributions

```python
n = [10, 100, 1000, 1000] #number of points
lamb = 2

for h in n:
    p = lamb / h

    # Binomial
    binomial =[]
    for i in range(0, h):
        binomial.append(statisticaltools.moeda(p, h))

    plt.hist(binomial, color = "darkviolet")

    #Poisson

    poisson = np.random.poisson(lamb, h)

    plt.hist(poisson, color = "aquamarine")
    plt.show()
```
